```python
# calculate the final price after discount for each product.
# calculate the total revenue for each product.
# for above q1 and q2, add both results as new columns to the dataframe.
# filter and display only the products where total revenue is greater than 500000
# calculate the total revenue of all products compiled.

import pandas as pd

data = {'product': ["Laptop", "Smartphone", "Tablet"],
        'unit_sold': [50, 150, 100],
        'price': [120000, 50000, 30000],
        'discount': [0.1, 0.2, 0.15]}

df = pd.DataFrame(data)

df['final_price'] = df['price'] * (1 - df['discount'])
df['total_revenue'] = df['final_price'] * df['unit_sold']

filtered_df = df[df['total_revenue'] > 500000]
print(filtered_df)
```
